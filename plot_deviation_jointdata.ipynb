```python
import plotly.graph_objs as go
```

```python
x = [0.0, 0.1, 0.25, 0.5]

y_baseline = [44.05, 42.38, 42.10, 40.47]
stddev_baseline = [3.07, 3.25, 2.80, 2.14]

y_sentencesim = [77.46, 77.16, 76.52, 76.02]
stddev_sentencesim = [1.61, 2.11, 1.67, 2.31]

def upper(y, stddev):
    return [y_i + stddev_i for y_i, stddev_i in zip(y, stddev)]

def lower(y, stddev):
    return [y_i - stddev_i for y_i, stddev_i in zip(y, stddev)]
```

```python
fig = go.Figure(
    [
        go.Scatter(
            name="Baseline",
            x=x,
            y=y_baseline,
            mode='lines+markers',
            line=dict(color='rgb(0,0,0)'),
        ),
        go.Scatter(
            x=x+x[::-1],
            y=upper(y_baseline, stddev_baseline)+lower(y_baseline, stddev_baseline)[::-1],
            fill='toself',
            fillcolor='rgba(0,0,0,0.2)',
            line=dict(color='rgba(255,255,255,0)'),
            showlegend=False
        ), 
         go.Scatter(
            name="Sentencesim CTS",
            x=x,
            y=y_sentencesim,
            mode='lines+markers',
            line=dict(color='rgb(19,174,240)')
        ),
        go.Scatter(
            x=x+x[::-1],
            y=upper(y_sentencesim, stddev_sentencesim)+lower(y_sentencesim, stddev_sentencesim)[::-1],
            fill='toself',
            fillcolor='rgba(0,0,0,0.2)',
            line=dict(color='rgba(255,255,255,0)'),
            showlegend=False
        ), 
    ]
)
```

```python
fig.update_layout(
    yaxis_title='Success (combined)',
    xaxis_title='% Noise',
    autosize=False,
    width=500,
    height=700,
)
fig.show()
```
